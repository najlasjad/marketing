```python
import pandas as pd
df = pd.read_csv('/Users/nauvalharyadi/Downloads/data cleaning/cleaned_2023_2024.csv', delimiter=';')

df.shape
```

```python
for col in df.columns:
    if df[col].dtype == 'O':
        df[col] = df[col].fillna('Unknown')
    else:
        df[col] = df[col].fillna(0)
```

```python
df_sorted = df.sort_values(by='isrankconfirm', ascending=False)
df_dedup = df_sorted.drop_duplicates(subset='email', keep='first')
```

```python
df_dedup.to_csv('realcleaned2023-2024.csv', index=False)
```
