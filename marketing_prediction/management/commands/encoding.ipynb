```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# 1. Load data yang sudah dibersihkan
df = pd.read_csv('../../../data/2022_clean_data.csv', sep=';')

# 2. Tambahkan fitur: asal dari Jabodetabek (kode provinsi: 31 = DKI Jakarta, 32 = Jawa Barat, 36 = Banten)
jabodetabek_prov = [31, 32, 36]
df['asal_dari_jabodetabek'] = df['iddataprovinces'].apply(lambda x: 1 if x in jabodetabek_prov else 0)

# 3. Identifikasi kolom kategori dengan jumlah unique value < 15
#    (asumsi ini adalah kolom kategori yang bisa di-encode)
categorical_cols = [col for col in df.columns if df[col].nunique() < 15 and df[col].dtype in ['int64', 'object']]

# 4. Pisahkan LabelEncoder dan OneHotEncoder:
#    - LabelEncoder: jika tipe data numerik (ordinal)
#    - OneHotEncoder: jika tipe data object (nominal)
ordinal_cols = [col for col in categorical_cols if df[col].dtype != 'object']
nominal_cols = [col for col in categorical_cols if df[col].dtype == 'object']

# 5. Terapkan LabelEncoder ke kolom ordinal
label_encoders = {}
for col in ordinal_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# 6. Terapkan OneHotEncoder via pd.get_dummies untuk kolom nominal
df = pd.get_dummies(df, columns=nominal_cols)

# 7. Simpan hasil encoding ke file baru
df.to_csv('../../../data/2022_encoded_data.csv', index=False)

# 8. Tampilkan 5 baris pertama
df.head()
```
